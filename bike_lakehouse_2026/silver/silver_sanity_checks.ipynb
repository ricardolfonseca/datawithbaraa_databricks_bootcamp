### Silver Layer Sanity Checks

This notebook serves as the ending point for the **Silver layer**, making sure all silver tables are available.

```python
expected_tables = 6         # Change if more tables are added

df = spark.sql("""
SELECT table_name
FROM information_schema.tables
WHERE table_name IN (
  'crm_customers',
  'crm_products',
  'crm_sales',
  'erp_customers',
  'erp_customer_location',
  'erp_product_category'
)
AND table_schema = 'silver'
AND table_catalog = 'workspace'
""")

row_count = df.count()
if row_count != expected_tables:              
    raise SystemError(f"Expected {expected_tables} rows, found {row_count}")
else:
    display(df)
    print(f"{expected_tables} tables found with success.")
```
