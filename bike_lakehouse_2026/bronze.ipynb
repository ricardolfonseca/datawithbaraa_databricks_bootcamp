### Ingesting Raw Data

We read all CSV files from the `bronze_config.py` path dictionary, ingesting them as raw data. Each file is loaded and written as a Delta table in the bronze layer, preserving the original data for further processing and analysis.

```python
from bronze_config import INGESTION_CONFIG
```

```python
%sql
USE CATALOG workspace;
USE SCHEMA bronze;
```

```python
try:
    for item in INGESTION_CONFIG:
        print(f"Ingesting {item['source']} → bronze.{item['table']}")

        df = (
            spark.read
                .option("header", "true")
                .option("inferSchema", "true")
                .csv(item["path"])
        )

        (
            df.write
            .mode("overwrite")
            .format("delta")
            .saveAsTable(f"workspace.bronze.{item['table']}")
        )
    print('Ingestion complete.')
    
except Exception as e:
    print(f"Error: {e}")   
```
