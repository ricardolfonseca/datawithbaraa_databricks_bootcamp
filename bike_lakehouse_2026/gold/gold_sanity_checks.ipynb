### Gold Layer Sanity Checks

This notebook serves as the ending point for the **Gold layer**, making sure all silver tables are available.

```python
expected_tables = 3         # Change if more tables are added

df = spark.sql("""
SELECT table_name
FROM information_schema.tables
WHERE table_name IN (
  'dim_customers',
  'dim_products',
  'fact_sales'
)
AND table_schema = 'gold'
AND table_catalog = 'workspace'
""")

row_count = df.count()
if row_count != expected_tables:              
    raise SystemError(f"Expected {expected_tables} rows, found {row_count}")
else:
    display(df)
    print(f"{expected_tables} tables found with success.")
```
